### os.listdir : 해당 디렉토리에 있는 폴더, 파일들의 리스트를 구해줌.

```python
import os 
def search_2(dirname):
    filenames=os.listdir(dirname) # os.listdir를 이용하면 해당 디렉토리에 있는 폴더들의 리스트를 구할 수 있음.
    for filename,_ in zip(filenames,range(5)): #zip 활용해서 5번만 for loop 돌게 하기
        print(filename)
        
search_2("c:/")
```

### os.path.join : dirname과 filename을 합쳐 주어서 진짜 filepath 처럼 보이게 해줌

```python
#import os

def search_3(dirname):
    filenames=os.listdir(dirname) #dirname 디렉토리에 있는 폴더들 구하기. 나는 dirname= c:/ 라고 함.
    for filename,_ in zip(filenames,range(5)): #filenames는 list 형식
        full_filename=os.path.join(dirname,filename) #os.path.join 이용해서 디렉토리 포함한 전체경로 구하기
        print(full_filename)
        
search_3("c:/")
```

### os.path.splitext : 확장자 앞에 있는 점(c:/abc.txt)을 기준으로 나누어 줌. c:/abc & .txt

```python
#import os

def search_4(dirname):
    filenames = os.listdir(dirname)
    for filename,_ in zip(filenames,range(5)):
        full_filename = os.path.join(dirname, filename)
        ext = os.path.splitext(full_filename) 
        print(ext)

search_4("c:/")
```

```python
#import os

def search_5(dirname):
    filenames = os.listdir(dirname)
    for filename in filenames:
        full_filename = os.path.join(dirname, filename)
        ext = os.path.splitext(full_filename)[-1]
        if ext == '.py': 
            print(full_filename)

search_5("c:/") # c드라이브에는 .py로 끝나는 파일이 없어서 아무것도 출력되지 않았음
```

### search를 두 번 쓰면 그 폴더 안에 계속 들어갈 수 있는 효과를 가짐 
- search('c:/') 하면 c:/에 있는 폴더, 파일들의 목록을 리스트로 만들 수 있음. 
- 이 중 하나의 폴더를 다시 잡아서 search를 쓰면 어떻게 될까?
- 예를 들어서 search('c:/BK Robot Journalism')을 쓰면 이 BK Robot Journalism 폴더 안에 들어있는 폴더, 파일들의 목록을 다 리스트로 만듦.
- 아래 search_6의 예처럼 nested loop 만들면 점화식처럼 'n이 있다면 n+1로 가라'는 느낌으로 무한히 진행됨 

```python
#import os

def search_6(dirname):
    try:
        filenames = os.listdir(dirname)
        for filename in filenames:
            full_filename = os.path.join(dirname, filename)
            if os.path.isdir(full_filename):
                search_6(full_filename) #재귀호출 - 자기 자신을 다시 호출하는 프로그래밍 기법
            else:
                ext = os.path.splitext(full_filename)[-1]
                if ext == '.py': 
                    print(full_filename)
    except PermissionError:
        pass

search_6("c:/")
```
